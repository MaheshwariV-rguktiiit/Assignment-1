2.Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line. 
 
 

```python
list1=[]
for n in range(2000, 3201):#Taken 2000 to 3201 because (both included)
    if (n%7==0) and (n%5!=0):
        list1.append(str(n))
print ("The numbers are\n",','.join(list1))
#taken example 2002 it is 2002/7 = 286 but not a multiple of 5 if we will divide it will give 2002/5 = 400.4 . easyily we can say last 1 digit is not multiple of 5 i.e not 5 or 0  
```

4.Write a Python program to find the volume of a sphere with diameter 12 cm.  
 
Formula: V=4/3 * π * r 3 

```python
import math
r = float(12/2) # converting diameter to radius d/2 .The radius is half the diameter so you have a radius of 12/2 = 6 cm.
pi = math.pi
V = (float(4/3)*pi*r**3)
print('Volume of a sphere with diameter 12cm =', V)
```

Create the below pattern using nested for loop in Python. 


```python
number_pattern=5
for x in range(number_pattern):
    for y in range(x):
        print ('* ', end="")
    print('')

for x in range(number_pattern,0,-1):
    for y in range(x):
        print('* ', end="")
    print('')
```
